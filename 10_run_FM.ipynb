## Run FM on hahn_2023_cortex_STAR_counts_TMMF_All_130_outliers_removed_44764genes_cleaned.csv

```r
expr <- fread("data/cleaned/ModSeed/hahn_2023_cortex_STAR_counts_TMMF_All_130_outliers_removed_44764genes_cleaned.csv", data.table=FALSE)

# For duplicate genes, choose row with highest mean expression

mean_expr <- data.frame(
    Index=1:nrow(expr), 
    Gene=expr[,1], 
    Expr=rowMeans(expr[,-1])
)

mean_expr <- mean_expr %>%
    group_by(Gene) %>%
    slice_max(Expr, with_ties=FALSE)

print(dim(mean_expr))

expr <- expr[mean_expr$Index,]

# Encountering errors when calculating distance in FM; problem seems related to some genes having 0 MAD (median absolute deviation)

# Filtering these in advance:

X <- as.matrix(expr[,-1])
mads <- apply(X, 1, mad, na.rm = TRUE)
bad  <- which(!is.finite(mads) | mads == 0)

length(bad)
head(expr[,1][bad])

mask <- mads > 0
expr <- expr[mask,]
```

```r
dim(expr)
```

```r
colnames(expr)[1] <- "Gene"

# Order samples by covariates of interest

sampleinfo <- read.csv("/mnt/lareaulab/reliscu/projects/NSF_GRFP/data/bulk/hahn_2023/cortex/hahn_2023_cortex_sampleinfo.csv")
sampleinfo <- sampleinfo %>%
    arrange(age, sex_y)
sampleinfo <- sampleinfo[sampleinfo$Run %in% colnames(expr),]
    
expr <- expr[, c(1, match(sampleinfo$Run, colnames(expr)[-1]) + 1)]
all.equal(sampleinfo$Run, colnames(expr)[-1])
```

```r
# Subset to genes in the top X percentile for generating modules

# subset <- NULL
prob <- .1
mean_expr <- apply(expr[,-1], 1, median)

subset_cutoff <- unname(quantile(mean_expr, prob))
print(paste("quantile(mean_expr, prob):", round(subset_cutoff, 3)))
subset <- mean_expr >= subset_cutoff
sum(subset)
```

```r
samplegroups <- as.factor(sampleinfo$age)
merge.param <- 0.93
signumvec <- c(.95,.94,.93,.92,.91,.9,.89,.88)
projectname <- paste0("hahn_2023_cortex_STAR_counts_TMMF_All_130_outliers_removed_44764genes_cleaned", "_mergeParam", merge.param, "_minSignum", min(signumvec))
```

```r
FindModules(
  projectname=projectname,
  expr=expr,
  geneinfo=1,
  sampleindex=2:ncol(expr),
  samplegroups=samplegroups,
  subset=subset,
  simMat=NULL,
  saveSimMat=FALSE,
  simType="Bicor",
  beta=1,
  overlapType="None",
  TOtype="signed",
  TOdenom="min",
  MIestimator="mi.mm",
  MIdisc="equalfreq",
  signumType="rel",
  iterate=TRUE,
  signumvec=signumvec, 
  minsizevec=c(3,4,6,8,10), 
  signum=NULL,
  minSize=NULL,
  merge.by="ME",
  merge.param=merge.param,
  export.merge.comp=T,
  ZNCcut=2,
  calcSW=FALSE,
  loadTree=FALSE,
  writeKME=TRUE,
  calcBigModStat=FALSE,
  writeModSnap=TRUE
)
```
